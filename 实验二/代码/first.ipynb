```python
import pandas as pd
import random
```

```python
df=pd.read_csv('./data/data3.csv')
```

```python
df.info()
#将数据按照C1进行升序
df.sort_values(by="C1",inplace=True)
df.head()
C1=df["C1"].to_list()
constitution=df["Constitution"].to_list()
for x in range(len(constitution)):
    if constitution[x]=="bad":
        constitution[x]=random.randint(0,59)
    elif constitution[x]=="general":
        constitution[x] =random.randint(60,75)
    elif constitution[x] == "good":
        constitution[x] =random.randint(76,85)
    else:
        constitution[x]=random.randint(86,100)


#引入散点图
import pyecharts.options as opts
from pyecharts.charts import Scatter
scatter=(
    Scatter()
    .add_xaxis(
        xaxis_data=C1
    )
    .add_yaxis(
        series_name="",
        y_axis=constitution,
        symbol_size=4,
        label_opts=opts.LabelOpts(is_show=False)
    )
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(type_="value"),
        yaxis_opts=opts.AxisOpts(type_="value"),
        title_opts=opts.TitleOpts(title="C1-体育图",pos_left="center")
    )
)
scatter.render_notebook()
```
